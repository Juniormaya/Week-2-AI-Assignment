```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
from io import StringIO # For creating dummy data

# --- Dummy Data for Demonstration ---
# In a real scenario, y_true and y_pred would come from your model's predictions
data = """num_antenatal_visits,previous_complications,age,anemia_status,hypertension_status,distance_to_clinic,education_level,high_risk_pregnancy
3,0,28,0,0,5,12,0
1,1,35,1,0,10,8,1
4,0,25,0,0,2,14,0
2,0,30,0,1,8,10,1
1,0,40,1,1,15,6,1
5,0,27,0,0,1,16,0
2,1,32,0,0,7,9,1
3,0,29,0,0,4,11,0
1,1,38,1,1,12,7,1
4,0,26,0,0,3,13,0
3,0,31,0,0,6,10,0
2,1,39,1,0,9,8,1
1,0,24,0,0,2,15,0
5,0,33,0,0,1,12,0
2,0,36,0,1,11,9,1
4,0,29,0,0,5,14,0
3,1,34,1,0,8,7,1
"""
df = pd.read_csv(StringIO(data))

X = df[['num_antenatal_visits', 'previous_complications', 'age', 'anemia_status', 'hypertension_status', 'distance_to_clinic', 'education_level']]
y = df['high_risk_pregnancy']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y) # stratify for balanced classes

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# --- Visualization ---
cm = confusion_matrix(y_test, y_pred, labels=[0, 1]) # Specify labels for clarity
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Low Risk', 'High Risk'])
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix for Maternal Health Risk Prediction')
plt.show()

# Normalized Confusion Matrix (often more insightful for imbalanced datasets)
# Calculate normalized confusion matrix (normalize by row)
cm_normalized = confusion_matrix(y_test, y_pred, labels=[0, 1], normalize='true')  # Normalize along rows
disp_normalized = ConfusionMatrixDisplay(confusion_matrix=cm_normalized, display_labels=['Low Risk', 'High Risk'])
disp_normalized.plot(cmap=plt.cm.Blues)
plt.title('Normalized Confusion Matrix (Row-wise)')
plt.show()
```
